```python
# Helper functions to generate data
def generate_random_date(start_year=2000, end_year=2024):
    start_date = datetime.strptime(f'{start_year}-01-01', '%Y-%m-%d').date()
    end_date = datetime.strptime(f'{end_year}-12-31', '%Y-%m-%d').date()
    return fake.date_between(start_date=start_date, end_date=end_date)

def get_unique_name(tribe_data, tribe):
    first_name = random.choice(tribe_data[tribe]['first_names'])
    last_name = random.choice(tribe_data[tribe]['last_names'])
    return first_name, last_name

def get_parent_name(tribe, religion, last_name):
    # Assign Christian or Islamic names based on religion
    if religion == "Christianity":
        parent_first_name = random.choice(['John', 'Mary', 'Grace', 'Paul', 'Joseph'])
    else:
        parent_first_name = random.choice(['Abdullahi', 'Fatimah', 'Amina', 'Umar'])
    
    return f'{parent_first_name} {last_name}'

# Number of students per tribe
num_students_per_tribe = 1000
num_students = num_students_per_tribe * len(tribe_data)  # 5 tribes
```

### Activities Data 

```python
# Number of students and activities
num_students = 5000
activity_categories = ['Academic', 'Sports', 'Extracurricular', 'Cultural']
frequency_choices = ['Daily', 'Weekly', 'Bi-weekly', 'Monthly']

# Example activities common in Nigerian senior secondary schools
academic_activities = ['Science Practical Sessions', 'Jets Club', 'Debate Club', 'Mathematics Competitions', 'Chemistry Lab Work']
sports_activities = ['Football', 'Basketball', 'Athletics', 'Chess', 'Scrabble', 'Volleyball']
extracurricular_activities = ['Drama Club', 'Press Club', 'Quiz Competition', 'Social Events Planning']
cultural_activities = ['Traditional Dance', 'Music Band', 'Cultural Day Rehearsals']
```

```python
# Function to randomly select an activity type and its category
def random_activity():
    category = random.choice(activity_categories)
    if category == 'Academic':
        return random.choice(academic_activities), category
    elif category == 'Sports':
        return random.choice(sports_activities), category
    elif category == 'Extracurricular':
        return random.choice(extracurricular_activities), category
    elif category == 'Cultural':
        return random.choice(cultural_activities), category
```

```python
# List to store activity data
activities_data = {
    'Activity_ID': [],
    'Student_ID': [],
    'Activity_Type': [],
    'Activity_Category': [],
    'Frequency_of_Participation': [],
    'Duration_per_Session (Hours)': [],
    'Impact_on_Performance': [],
    'Teacher_Supervisor': [],
    'Parental_Support': []
}
```

```python

# For each student, generate between 1 and 5 activities
for student_id in students_df['Student_ID']:
    num_activities = random.randint(1, 5)
    for _ in range(num_activities):
        activity_id = fake.unique.random_int(min=100000, max=999999)
        activity_type, category = random_activity()
        frequency = random.choice(frequency_choices)
        duration = round(random.uniform(1, 3), 2)
        impact = random.choice(['Positive', 'Negative', 'Neutral'])
        supervisor = fake.name()
        parental_support = random.choice(['Yes', 'No'])
        
        # Append data for each activity
        activities_data['Activity_ID'].append(activity_id)
        activities_data['Student_ID'].append(student_id)
        activities_data['Activity_Type'].append(activity_type)
        activities_data['Activity_Category'].append(category)
        activities_data['Frequency_of_Participation'].append(frequency)
        activities_data['Duration_per_Session (Hours)'].append(duration)
        activities_data['Impact_on_Performance'].append(impact)
        activities_data['Teacher_Supervisor'].append(supervisor)
        activities_data['Parental_Support'].append(parental_support)
```

```python
# Convert the dictionary to a pandas DataFrame
activities_df = pd.DataFrame(activities_data)

# Display the first few rows to verify
activities_df.head()
```

```python
import random
import pandas as pd

# Assuming the parents_df is already created with 'Student_ID'
# Extracting Student_IDs from the parents_df
student_ids = students_df['Student_ID'].tolist()

# Number of students (should be the same as in parents_data)
num_students = len(student_ids)

# Function to generate random student-specific variables
def generate_student_data(student_ids):
    teacher_experience = [random.randint(5, 20) for _ in range(num_students)]  # Experience of teachers in years
    return {
        "Student_ID": student_ids,
        "Teacher_Student_Ratio": [random.uniform(15, 25)] * num_students,  # Ratio of teachers to students
        "Average_Teacher_Experience_Years": teacher_experience,  # Experience of teachers in years
        "Average_Class_Size": [random.randint(20, 35)] * num_students,  # Average number of students per class
        "School_Funding_Per_Student": [random.randint(10000, 30000) for _ in range(num_students)],  # Varying funding per student in USD
        "School_Extracurricular_Activities": [random.choice(['Sports', 'Arts', 'Both', 'None']) for _ in range(num_students)],  # Extracurriculars
        "Parental_Involvement_Score": [random.uniform(5.0, 10.0) for _ in range(num_students)],  # Parental involvement on a scale of 1 to 10
        "School_Facilities_Rating": [random.uniform(6.0, 10.0 - (exp / 20)) for exp in teacher_experience],  # Facilities rating influenced by teacher experience
        "Internet_Access_In_School": [random.choice([True, False]) for _ in range(num_students)],  # Whether school has internet access
        "School_Distance_From_Home_km": [random.uniform(0.5, 5.0) for _ in range(num_students)],  # Distance from home in km
        "Student_Attendance_Rate": [random.uniform(80, 100) for _ in range(num_students)],  # Attendance rate as percentage
        "Disciplinary_Actions_Taken": [random.randint(0, 3) for _ in range(num_students)],  # Disciplinary actions taken
        "Student_Performance_Score": [random.uniform(20, 100) for _ in range(num_students)],  # Performance score of student (20 to 100)
    }

# Generate school data for one school
school_data = generate_student_data(student_ids)

# Convert to DataFrame
school_df = pd.DataFrame(school_data)
```
