```python
activities_data['Activity_Type'] = activities_data['Activity_Type'].astype('category')

average_duration = activities_data.groupby('Activity_Type')['Duration_per_Session (Hours)'].mean().reset_index()

average_duration = average_duration.sort_values('Activity_Type')

plt.figure(figsize=(18, 6))
sns.lineplot(data=average_duration, x='Activity_Type', y='Duration_per_Session (Hours)', marker='o', color='orange')

plt.title('Average Duration per Session by Activity Type')
plt.xlabel('Activity Type')
plt.ylabel('Average Duration per Session (Hours)')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
